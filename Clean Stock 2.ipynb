```python
import pandas as pd
```

```python
df = pd.read_csv('全量数据/股本变化.csv')
```

```python
df['截止日期'] = pd.to_datetime(df['截止日期'], format='%d/%m/%Y %H:%M:%S')
df['截止日期'] = df['截止日期'].dt.strftime('%Y-%m-%d')
```

```python
df['证券代码'] = df['证券代码'].astype(str).apply(lambda x: x.zfill(6))
```

```python
df = df[~df['证券代码'].astype(str).str.startswith(('4', '8', 'X', '9', 'S'))]
```

```python
def modify_code(code):
    if code.startswith('0') or code.startswith('3'):
        return code + '.SZ'
    elif code.startswith('6'):
        return code + '.SH'
    else:
        return code  # 如果不符合条件，保持原样

df['证券代码'] = df['证券代码'].astype(str).apply(modify_code)
```

```python
df = df.drop(df.columns[[1, 4, 5]], axis=1)
df
```

```python
df_1 = df.pivot(index='截止日期', columns='证券代码', values='总股本')
```

```python
df_1.index.name = None
df_1.columns.name = None
```

```python
df_1
```

```python
df_1.index = pd.to_datetime(df_1.index)
df_1 = df_1.resample('D').asfreq()
```

```python
df_1
```

```python
df_1 = df_1.ffill()
```

```python
df_1 = df_1.fillna(0)
```

```python
df_1
```

```python
df_1.to_csv('历史总股本.csv')
```
